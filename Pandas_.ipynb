# Hang Man Game

```python
import random
import os 
def play_again():
    question = "Do You wanna pplay again?! y = YES, n= No\n"
    play_again = input(question)
    while play_again.lower() not in ['y','n','yes','no']:
        print('Inavlid input\n')
        play_again = input(question)
    if play_again.lower() =='y' or play_again.lower() == 'yes' :
        return True
    else:
        return False
    
```

```python
def hangman(word):
    display = '_'*len(word)
    limit = 5
    letter = list(word)
    guessed = []
    count = 0
    g = f"Hangman Word : {display} Enter Your Guess : \t".strip()
    while count < limit :
        guess = input(g)
        while len(guess) ==0 or len(guess) > 1 :
            print("Invalid Input : I need 1 letter\n")
            guess = input(g)
            
        if guess in guessed :
            print('Oops You already guessed that!\n')
            continue
            
        if guess in letter:
            letter.remove(guess)
            index = word.find(guess)
            display = display[:index] + guess + display[index+1:]
        else :
            guessed.append(guess)
            count +=1
            if count == 1 :
                print('   _____ \n'
                    '  |      \n'
                    '  |      \n'
                    '  |      \n'
                    '  |      \n'
                    '  |      \n'
                    '  |      \n'
                    '__|__\n')
                    
                print(f'Wrong guess: {limit - count} guesses remaining\n')
            elif count == 2 :
                print('   _____ \n'
                    '  |       |\n'
                    '  |       |\n'
                    '  |      \n'
                    '  |      \n'
                    '  |      \n'
                    '  |      \n'
                    '__|__\n')
                
                print(f'Wrong guess: {limit - count} guesses remaining\n')
            elif count == 3:
                print('   _____ \n'
                    '  |       |\n'
                    '  |       |\n'
                    '  |       |\n'
                    '  |      \n'
                    '  |      \n'
                    '  |      \n'
                    '__|__\n')
                
                print(f'Wrong guess: {limit - count} guesses remaining\n')
            elif count == 4:
                print('   _____ \n'
                    '  |       |\n'
                    '  |       |\n'
                    '  |       |\n'
                    '  |       O\n'
                    '  |      \n'
                    '  |      \n'
                    '__|__\n')
                
                print(f'Wrong guess: {limit - count} guesses remaining\n')
            elif count == 5:
                print('   _____ \n'
                    '  |     | \n'
                    '  |     | \n'
                    '  |     | \n'
                    '  |     O \n'
                    '  |    /|\ \n'
                    '  |    / \ \n'
                    '__|__\n')
                
                print('YOU HAVE BEEN HANGED !!') 
                print(f'The word was {word}')
                
        if display == word :
            print(f'Congratulations You Won ! You Guessed {word} Correctly!')
            break
                


                
```

```python
def play_hangman():
    d = '*'*20
    print(f'{d}Welcome To Hangman{d}')
    name = input("Please Enter Your Name\n")
    print(f'Hello {name}! Best of Luck')
    print(f'The game is About To start in 5')
    print('Rules : You have 5 Wrong guesses')
    words_to_guess = ['Mohamed','Samar','Youssef','Omar']
    play = True
    while play :
        word = random.choice(words_to_guess).lower()
        hangman(word)
        play = play_again()
    print("Tanks For Playing! See you again !")
    exit()
if __name__ == '__main__':
    play_hangman()
```
